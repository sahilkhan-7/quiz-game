```python
# Importing necessary libraries
import pandas as pd
import mysql.connector as connector
```

```python
# Defining the database configuration

config = {
    'user': 'root',
    'password': 'sahilkhan@7824',
    'host': 'localhost',
    'database': 'quizgame'
}
```

```python
# SQL queries to create tables

data_table_query = '''
CREATE TABLE quizdata(
    sno INT AUTO_INCREMENT PRIMARY KEY, 
    question VARCHAR(255), 
    optionA VARCHAR(100),
    optionB VARCHAR(100),
    optionC VARCHAR(100),
    optionD VARCHAR(100),
    correct_ans VARCHAR(1) NOT NULL
)'''

login_info_query = '''
CREATE TABLE user_info(
    fullname VARCHAR(50) NOT NULL,
    phone VARCHAR(10),
    email VARCHAR(50) NOT NULL,
    username VARCHAR(50) PRIMARY KEY,
    password VARCHAR(50) NOT NULL,
    score INT DEFAULT 0
)'''

# Creating a list of queries, to execute them in a loop
queries = [data_table_query, login_info_query]


# Establishing connection with the database, and checking if the connection is successful
try:
    connection = connector.connect(**config)
    if connection.is_connected():
        print("Connected to MySQL database")
        
    for i in range(2):
        cursor = connection.cursor()
        cursor.execute(queries[i])
        connection.commit()
        print("Table created successfully.")
    
finally:
    if connection.is_connected():
        cursor.close()
        connection.close()
        print("MySQL connection is closed")
```

```python
# Loading the dataset
df = pd.read_csv('quiz_data.csv')
```

```python
df
```

```python
# Converting the dataframe numpy object to python object, so that it can be inserted into the database
df = df.astype(object).where(pd.notnull(df), None)
```

```python
df
```

```python
# Establishing connection with the database, and checking if the connection is successful

try:
    connection = connector.connect(**config)
    if connection.is_connected():
        print("Connected to MySQL database")
        
        cursor = connection.cursor()

        # Insert df into the quizdata table
        insert_query = '''
        INSERT INTO quizdata (sno, question, optionA, optionB, optionC, optionD, correct_ans)
        VALUES (%s, %s, %s, %s, %s, %s, %s)
        '''
        for row in df.itertuples(index=False):
            cursor.execute(insert_query, tuple(row))
        
        connection.commit()
        print("Data inserted successfully.")

finally:
    if connection.is_connected():
        cursor.close()
        connection.close()
        print("MySQL connection is closed")
```
